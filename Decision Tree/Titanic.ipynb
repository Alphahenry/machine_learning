#### The sinking of the Titanic is one of the most horrific  event in history.It led to the loss of  over 1000 lives. Reseachers  have since been piqued by the reasons behind the tragic loss of lives despite being well equiped with safety equipments.

#### In this  research project I am going to use decision  tree  to determine variables that influenced the chances of survival in the titanic.

```python
#load standard libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
df = pd.read_csv("titanic.csv")
df.head()
```

```python
#calculating correlation between survival and number of siblings
corr = df[['Survived', "SibSp"]].corr()
corr
```

```python
corr = df[['Survived', "Fare"]].corr()
corr
```

```python
features = df[["Pclass", "Sex", "Age","Fare" ]]
```

```python
features.head(2)
```

```python
target = df['Survived']
```

```python
features.isnull()
```

```python
mean_value = features['Age'].mean()
mean_value
```

```python
print(features["Sex"].dtype)
```

```python
print(features["Age"].dtype)
```

```python
features["Age"]
```

```python

features['Age'].fillna(value=mean_value, inplace=True)
```

```python
features["Age"].isnull
```

```python
features.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
sex_enc = LabelEncoder()
```

```python
features["Sex"] = sex_enc.fit_transform(features["Sex"])
```

```python
features
```

```python
features.head()
```

```python
X = features
```

```python
y= target
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test =  train_test_split(X, y, train_size=0.2)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```
